```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn import linear_model
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
df = pd.read_csv('paddyph1.csv')
df
```

```python
model = preprocessing.LabelEncoder()
```

```python
df['Taluks']=model.fit_transform(df['Taluks'])
df['Soil']=model.fit_transform(df['Soil'])
df['Paddy variety']=model.fit_transform(df['Paddy variety'])
df['Season']=model.fit_transform(df['Season'])
df
```

```python
x = df.drop(['Yield'],axis= 1)
y = df['Yield']
```

```python
xtrain,xtest,ytrain,ytest = sklearn.model_selection.train_test_split(x,y,train_size = 0.7)
print(xtrain.shape)
print(ytrain.shape)
```

```python
mlr = LinearRegression()
mlr.fit(xtrain,ytrain)
```

```python
print(f'Correlation Coefficient - {mlr.coef_}')
print(f'Intercept - {mlr.intercept_}')
```

```python
mlr_pred = mlr.predict(xtest)
a = mlr.score(xtest,ytest)
```

```python
ls = Lasso()
ls.fit(xtrain,ytrain)
```

```python
ls_pred = ls.predict(xtest)
b = ls.score(xtest,ytest)
```

```python
rg = Ridge()
rg.fit(xtrain,ytrain)
```

```python
rg_pred = rg.predict(xtest)
c = rg.score(xtest,ytest)
#Taluks	Soil	Paddy variety	Rainfall		Temperature Nitrogen	Soil ph	Duration	Season	Yield
```
